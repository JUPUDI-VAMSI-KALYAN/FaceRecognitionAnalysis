```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import matplotlib.image as matimg
```

```python
# OPencv- BGR returns numpy array
img_cv = cv2.imread('./data/flying_bird.jpg')
# Matplotlib - RGB returns numpy array
img_mat = matimg.imread('./data/flying_bird.jpg')
# PIL - RGB - returns object
img_pil = Image.open('./data/flying_bird.jpg')
```

```python
# Opencv
print(img_cv)
```

```python
#matplotlib
print(img_mat)
```

```python
# PIL
print(img_pil)
```

```python
img_pil
```

```python
# converting pil object to array
img_pil_arr = np.array(img_pil)
```

```python
img_pil_arr
```

# Display image

```python
# display image using opencv
cv2.imshow('Opencv BGR',img_cv)
cv2.imshow("Matplotlib RGB",img_mat)
cv2.imshow("PIL image",img_pil_arr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# display image using matpltlib
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_cv)
plt.axis("off")
plt.title("OPencv Image")

plt.subplot(1,3,2)
plt.imshow(img_mat)
plt.axis("off")
plt.title("Matplotlib Image")

plt.subplot(1,3,3)
plt.imshow(img_pil_arr)
plt.axis("off")
plt.title("PIL Image")
plt.show()
```

# COLOR Space

- color conversion
- Gray Scale

```python
img_cv.shape
```

```python
# numpy slicing
b = img_cv[:,:,0]
g = img_cv[:,:,1]
r = img_cv[:,:,2]
```

```python
cv2.imshow('Blue Channel',b)
cv2.imshow('Green Channel',g)
cv2.imshow('red Channel',r)

cv2.imshow("Original",img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Opencv split channel
b,g,r = cv2.split(img_cv)
```

```python
cv2.imshow('Blue Channel',b)
cv2.imshow('Green Channel',g)
cv2.imshow('red Channel',r)

cv2.imshow("Original",img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# converting color
img_cv_rgb = cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cv_rgb)
plt.show()
```

Gray Scale

```python
img_gray = cv2.cvtColor(img_cv,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray,cmap='gray')
plt.show()
```
